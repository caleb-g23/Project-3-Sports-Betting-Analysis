```python
# Dependencies
import requests
import csv
import json
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
#read the csvs file into a data frame
all_state_data= Path("Resources/all_state_market_data_FINALSET.csv")
all_state_df= pd.read_csv(all_state_data)
all_state_df
```

```python
all_state_df.dtypes
```

```python
#Change the date column to datetime format
all_state_df=all_state_df.astype({'date':'datetime64[D]'})
all_state_df
```

```python
#Change handle, revenue and taxes to dollar format
all_state_df['handle']=all_state_df['handle'].astype(float).map('${:,.2f}'.format)
all_state_df['revenue']=all_state_df['revenue'].astype(float).map('${:,.2f}'.format)
all_state_df['taxes']=all_state_df['taxes'].astype(float).map('${:,.2f}'.format)

#all_state_df
```

```python
#Change the hold column to show as a percentage
all_state_df['hold']=all_state_df['hold'].apply(lambda x: "{:.2%}".format(x))
#all_state_df
```

```python
#Import state and abbreviation csv
state_data=Path('Resources/state_list_REFERENCE.csv')
state_df=pd.read_csv(state_data)
#state_df
```

```python
#Merge the Datasets on state_abbreviation
all_state_merge=pd.merge(all_state_df,state_df, on='state_abbr', how='inner')
all_state_merge.head()
```

```python
#Import state and population csv
population_data=Path('Resources/state_pop.csv')
population_df=pd.read_csv(population_data)
#population_df
```

```python
#Merge the Datasets on state_abbreviation
all_state_pop=pd.merge(all_state_merge,population_df, on='state', how='inner')
all_state_pop.head()
```

```python
#Changing the population column to a float value
all_state_pop['population total 18+']=all_state_pop['population total 18+'].astype(float)
```

```python
#Add a new column to show the the average reveune per person per state
all_state_pop['Revenue Per Person']= all_state_pop['revenue']/ all_state_pop['population total 18+']
```

```python
#reorder columns to start with state
order=['state','state_abbr', 'date', 'handle', 'revenue','hold','taxes','population total 18+', 'Revenue Per Person']
all_state_pop=all_state_pop[order]
#all_state_pop
```

```python
#Change the hold column to show as a percentage
all_state_pop['hold']=all_state_pop['hold'].apply(lambda x: "{:.2%}".format(x))
#all_state_pop
```

```python
data= all_state_pop

# Create a dictionary to store revenue data for each state
state_revenue = {}

for row in data:
    state = row['state']
    revenue = row['Revenue Per Person']
    if state in state_revenue:
        state_revenue[state].append(revenue)
    else:
        state_revenue[state] = [revenue]

# Extract unique states and their corresponding revenue data
states = list(state_revenue.keys())
revenue_data = list(state_revenue.values())

# Create a line plot
plt.figure(figsize=(12, 8))

for i, state in enumerate(states):
    plt.plot(revenue_data[i], label=state)

plt.xlabel('Month')
plt.ylabel('Revenue Per Person')
plt.title('Revenue Per Person by State and Month')
plt.legend()
plt.xticks(range(len(revenue_data[0])), data['date'], rotation=45)  
plt.tight_layout()

plt.show()
```

```python
# Function to convert a CSV to JSON
# Takes the file paths as arguments
def make_json(csvFilePath, jsonFilePath):
     
    # create a dictionary
    data = {}
     
    # Open a csv reader called DictReader
    with open(csvFilePath, encoding='utf-8') as csvf:
        csvReader = csv.DictReader(csvf)
         
        # Convert each row into a dictionary
        # and add it to data
        for rows in csvReader:
             
            #primary key
            key = rows['']
            data[key] = rows
 
    # Open a json writer, and use the json.dumps()
    # function to dump data
    with open(jsonFilePath, 'w', encoding='utf-8') as jsonf:
        jsonf.write(json.dumps(data, indent=4))
```
